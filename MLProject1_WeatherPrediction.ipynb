```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('seattle-weather.csv')
df
```

```python
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
Lb= LabelEncoder()
y=Lb.fit_transform(df['weather'])
df['weather'].unique()
print(y)
```

```python
independent_variables = ['precipitation' , 'temp_max', 'temp_min', 'wind']

def normalize (df, independent_variables):
    for i in independent_variables:
        df[i]=df[i]/df[i].max()
normalize (df, independent_variables)
df
```

```python
df1=df.drop('date', axis=1)
df1
```

```python
x = df1.drop("weather", axis = 1)
print (x)
print (y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
print (x_train)
print(x_test)
print(y_train)
print (y_test)
```

```python
from xgboost import XGBClassifier
```

```python
xg = XGBClassifier()
xg.fit(x_train, y_train)
```

```python
xg.get_params()
```

```python
y_pred = xg.predict(x_test)
print (y_pred)
print (y_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
grid = {'learning_rate': [0.1,1, 0.01, 0.001], 'gamma':[0,1,10,100]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = GridSearchCV(XGBClassifier(), grid, cv=10, verbose=2)
```

```python
model.fit(x_train, y_train)
```

```python
grid_pred = model.predict(x_test)
print(classification_report(y_test, grid_pred))
print(accuracy_score(y_test,grid_pred))
```

```python
print(model.best_estimator_)
```
